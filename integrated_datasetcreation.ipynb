# Creating Rna-Drug_response (single omics representation)

```python
new_gdsc= gdsc_ic50[['COSMIC_ID', 'DRUG_ID','MIN_CONC','MAX_CONC', 'LN_IC50', 'CELL_LINE_NAME']] 
```

```python
rna_encoded
```

```python
response_new = pd.merge(new_gdsc, rna_encoded, on='CELL_LINE_NAME')
```

```python
response_new
```

```python
full_data = pd.merge(response_new, drug_encoded, on='DRUG_ID')
```

```python
full_data
```

```python
catogorical_col = []
numeric_col =[]
for i in full_data.columns:
    if(full_data[i].dtype == np.float64):
        numeric_col.append(i)
    elif (full_data[i].dtype == np.int64):
        numeric_col.append(i)
    elif (full_data[i].dtype == np.float16):
        numeric_col.append(i)
    
    else:
        print(i ,':',len(full_data[i].unique()) )
        
        catogorical_col.append(i)
```

```python
from google.colab import files
full_data.to_csv('Singleomics_Integrated_data_rna.csv') 
files.download('Singleomics_Integrated_data_rna.csv')
```
